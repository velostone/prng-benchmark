```python
import matplotlib.pyplot as plt
import numpy as np

SIZE = 10
```

```python
filename = '..\\results\\lattice-test\\mcg59_10_double'
pic = np.loadtxt(filename + '.txt')
image_data = np.array(pic).reshape(SIZE, SIZE)

fig = plt.figure(figsize=(20, 20), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.imshow(image_data, cmap='cividis', interpolation='nearest')
ax.axis('off')

fig.savefig(filename + '.png', dpi=100, bbox_inches='tight', pad_inches=0)
```
